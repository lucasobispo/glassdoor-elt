```python
import pandas as pd
import os
import numpy as np
from unidecode import unidecode
import re
```

```python
# funcao que retonar todos os arquivos de uma pasta em um dataframe
def filesToDataframe (path,sep, encoding):
    lista_df = []
    for file in os.listdir(path):
        path_file = os.path.join(path, file)
        if os.stat(path_file).st_size > 0:
            df_temp = pd.read_csv(path_file,sep=sep, encoding=encoding)
            lista_df.append(df_temp)
    df_final = pd.concat(lista_df, ignore_index=True)
    return df_final
```

```python
glassdoor_path = '/content/drive/MyDrive/Colab Notebooks/Pos_ingestao/Dados/Empregados'
df_glassdoor = filesToDataframe(glassdoor_path, '|' ,'utf8')
df_glassdoor.head(n=5)
```

```python
reclamacoes_path = '/content/drive/MyDrive/Colab Notebooks/Pos_ingestao/Dados/Reclamacoes'
df_reclamacoes = filesToDataframe(reclamacoes_path, ';' ,'iso-8859-1')
df_reclamacoes.head(n=5)
```

```python
banco_path = '/content/drive/MyDrive/Colab Notebooks/Pos_ingestao/Dados/Bancos'
df_banco = filesToDataframe(banco_path, '\t' ,'iso-8859-1')
df_banco.head(n=10)
```

```python
# Drop linhas duplicadas
df_glassdoor = df_glassdoor.drop_duplicates()
df_reclamacoes = df_reclamacoes.drop_duplicates()
df_banco = df_banco.drop_duplicates()
```

```python
# deixando apenas as colunas relevenates
df_glassdoor = df_glassdoor[['Nome','Geral', 'Remuneração e benefícios']]
df_reclamacoes = df_reclamacoes[['Instituição financeira','CNPJ IF','Índice','Quantidade total de reclamações','Quantidade total de clientes  CCS e SCR']]
```

```python
# funcao para transformar " " em vazio e converter os campos para numeros
def convertCol(cols_to_fix, df):
    for col in cols_to_fix:
        df[col] = df[col].replace({"": np.nan, " ": np.nan})
    for col in cols_to_fix:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    return df
```

```python
df_reclamacoes = convertCol(['Índice','Quantidade total de reclamações','Quantidade total de clientes  CCS e SCR'],df_reclamacoes)
# calculando as metricas de reclamacoes
df_reclamacoes = df_reclamacoes.groupby(['Instituição financeira', 'CNPJ IF']).agg({
    'Índice': 'mean',
    'Quantidade total de reclamações': 'sum',
    'Quantidade total de clientes  CCS e SCR': 'sum'
}).reset_index()
df_reclamacoes.head(n=5)
```

```python
df_glassdoor = convertCol(['Geral','Remuneração e benefícios'],df_glassdoor)
df_glassdoor = df_glassdoor.groupby(['Nome']).agg({
    'Geral': 'mean',
    'Remuneração e benefícios': 'mean'
}).reset_index()
df_glassdoor.head(n=5)
```

```python
# funcao para padronizar os nomes do banco, remover acentos, remover alguma palavra especifica, remover espacos em brnaco no comeco e final
# e deixar tudo maisculo
def standardize_column(df, column_name,word_to_remove=None ):
    # Remover acentos
    df[column_name] = df[column_name].apply(lambda x: unidecode(str(x)) if pd.notnull(x) else x)

    # Remover a palavra
    if word_to_remove != None:
         # Escapar caracteres especiais na palavra a ser removida
        escaped_word = re.escape(word_to_remove)
        pattern = r"\s*" + escaped_word + r"\s*"
        df[column_name] = df[column_name].str.replace(pattern, "", case=False, regex=True)

    # Remover espaços vazios no início e no final da string
    df[column_name] = df[column_name].str.strip()

    # Transformar tudo em maiúsculo
    df[column_name] = df[column_name].str.upper()

    return df
```

```python
# padronizando o nome da Instituição financeira na base de reclamacaoe e removendo a string (conglomerado)
df_reclamacoes = standardize_column(df_reclamacoes, 'Instituição financeira', '(conglomerado)')
df_reclamacoes.head(n=5)
```

```python
# padronizando o nome na base de bancos e removendo a string - PRUDENCIAL
df_banco = standardize_column(df_banco, 'Nome', '- PRUDENCIAL')
df_banco.head(n=5)
```

```python
# Inner Join da base do glassdoor com a de reclamacao
df_glassdoor_reclmacao = pd.merge(df_reclamacoes, df_glassdoor, left_on='Instituição financeira', right_on='Nome', how='inner')
df_glassdoor_reclmacao.head(n=50)
```

```python
# removendo a coluna Nome
df_glassdoor_reclmacao = df_glassdoor_reclmacao.drop(columns=['Nome'])
```

```python
# Inner Join com a base de bancos
df_final = pd.merge(df_glassdoor_reclmacao, df_banco, left_on='Instituição financeira', right_on='Nome', how='inner')
df_final.head(n=50)
```

```python
df_final = df_final.drop(columns=['Nome', 'CNPJ IF'])
df_final = df_final.rename(columns={
    'Instituição financeira': 'nome_do_banco',
    'CNPJ': 'cnpj',
    'Segmento': 'classificacao_do_banco',
    'Quantidade total de clientes  CCS e SCR':'quantidade_clientes_bancos',
    'Índice':'indice_reclamacoes',
    'Quantidade total de reclamações':'quantidade_reclamacoes',
    'Geral':'indice_satisfacao_funcionarios_bancos',
    'Remuneração e benefícios':'indice_satisfacao_salarios_funcionarios_bancos'
})
df_final.head(n=50)
```

```python
# colocando o resultado final em um csv
df_final.to_csv('tabela_final.csv')
```
